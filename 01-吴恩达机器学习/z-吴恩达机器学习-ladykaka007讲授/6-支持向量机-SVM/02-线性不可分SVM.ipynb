## 线性不可分支持向量机
## 任务：使用高斯核函数解决线性不可分问题，并观察gamma取值对模型复杂度的影响
## 数据集：data/ex6data2.mat

```python
# 线性不可分支持向量机

import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
data = sio.loadmat('./data/ex6data2.mat')
```

```python
data.keys()
```

```python
X,y = data['X'],data['y']
X.shape,y.shape
```

```python
def plot_data():
    plt.scatter(X[:,0],X[:,1],c = y.flatten(), cmap ='jet')
    plt.xlabel('x1')
    plt.ylabel('y1')
```

```python
plot_data()
```

```python
# 改变gamma 值，看性能变化    1  和  1000

svc1 = SVC(C=1,kernel='rbf',gamma=1000)   #  gamma 越大 复杂度越高
svc1.fit(X,y.flatten())
```

```python
svc1.score(X,y.flatten())
```

```python
def plot_boundary(model):
    x_min,x_max = 0,1
    y_min,y_max = 0.4,1
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,500),
                       np.linspace(y_min,y_max,500))
    z = model.predict(np.c_[xx.flatten(),yy.flatten()])
    
    zz = z.reshape(xx.shape)
    plt.contour(xx,yy,zz)
```

```python
plot_boundary(svc1)
plot_data()
```
