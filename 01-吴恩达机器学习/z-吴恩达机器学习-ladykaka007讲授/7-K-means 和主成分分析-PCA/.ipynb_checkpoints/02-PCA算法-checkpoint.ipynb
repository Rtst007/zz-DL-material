## 案例1：使用PCA进行二维数据的降维
## 数据集：data/ex7data1.ma

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
mat = sio.loadmat('./data/ex7data1.mat')
X = mat['X']
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

## 1. 对X去均值化

```python
X_demean = X - np.mean(X,axis=0)
plt.scatter(X_demean[:,0],X_demean[:,1])
plt.show()
```

## 2.计算协方差矩阵

```python
C = X_demean.T@X_demean / len(X)
C
```

## 3.计算特征值，特征向量

```python
U,S,V = np.linalg.svd(C)
```

```python
S
```

```python
U
```

```python
U1 = U[:,0]  # 取出 U1  的第一列
U1
```

## 4.实现降维

```python
X_reduction = X_demean@U1
```

```python
plt.figure(figsize=(7,7))   # 设置图像 为正方形， 所以 才能看出黑线和红线是垂直的
plt.scatter(X_demean[:,0],X_demean[:,1])
plt.plot(  [0,U1[0]],  [0,U1[1]],  c='r' )
plt.plot( [0,U[:,1][0]],  [0,U[:,1][1]],  c='k' )
plt.show()
```

## 5.还原数据

```python
X_restore = X_reduction.reshape(50,1)@U1.reshape(1,2) + np.mean(X,axis=0)
```

```python
X_reduction.shape,U1.shape
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(X_restore[:,0],X_restore[:,1])
plt.show()
```
